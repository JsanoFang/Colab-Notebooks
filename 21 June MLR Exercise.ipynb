```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
diabetes= load_diabetes()
x, y = load_diabetes(return_X_y=True)

diabetes_df = pd.DataFrame(diabetes.data,columns = diabetes.feature_names)
diabetes_df
```

```python
corr = diabetes_df.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.05, random_state =5)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(x_train_sc,y_train)
r2 = model.score(x_test_sc,y_test)
print("score %.4f" %r2)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error

model2 = SGDRegressor(random_state=1)
model2.fit(x_train_sc,y_train)

r2 = model.score(x_test_sc,y_test)
print("score %.4f"%r2)

y_pred = model2.predict(x_test_sc)

MSE = mean_squared_error(y_test,y_pred)
print("mse %.5f" %MSE )
print("coef", model2.coef_)
print("y intercept", model2.intercept_)
```

```python
from sklearn.model_selection import GridSearchCV

parameters={'loss':('squared_loss','huber'),'penalty':('l2','l1','elasticnet'),
            'alpha':[1e-10,1e-8,1e-6,1e-4,1e-2,1e-1,1],'max_iter':[4000,4500,5000,5500,6000,6500,7000]}

model3 = SGDRegressor(random_state=10)
clf = GridSearchCV(model3,parameters)
clf.fit(x_train_sc,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.best_estimator_.coef_
```

```python
clf.best_estimator_.intercept_
```

```python
best_model = clf.best_estimator_
best_model.score(x_test_sc, y_test)
```
