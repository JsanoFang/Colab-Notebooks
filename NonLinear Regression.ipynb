```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#import dataset (generate data for non-linear regression)
# x - data, y = quardratic equation
x = np.array(7* np.random.rand(100,1)-3)
```

```python
#y = 13x^2 +2x+7
y = 13*x*x + 2*x + 7
```

```python
x1 = x.reshape(-1,1)
```

```python
#data points
plt.scatter(x,y,s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non Linear Data')
plt.show()
```

```python
#polynomial degree 2 ( adding nonlinear features) to cater to the polynomial relationship
#degree 2 because x^2 was used
polyfeatures = PolynomialFeatures(degree=2,include_bias=False)
xpoly = polyfeatures.fit_transform(x1)
x[3]
```

```python
xpoly[3]
```

```python
#use xpoly instead of x in order to cater to nonlinear regression
model = LinearRegression()
model.fit(xpoly,y)

print('coefficients of x are', model.coef_)
print('intercept value is', model.intercept_)

#predict
ydeg2 = model.predict(xpoly)

#model evaluation
print('MSE of polyregression model: %.2f' % mean_squared_error(y,ydeg2))
print('r2 score of non linear model:', model.score(xpoly,y))
```

```python
x_new = np.linspace(-3,4,100).reshape(100,1)
x_newpoly = polyfeatures.transform(x_new)
y_new = model.predict(x_newpoly)
plt.plot(x,y,"b.")
plt.plot(x_new,y_new,"r.",linewidth = 2,label = "Predictions")
plt.xlabel("$x_1$",fontsize=10)
plt.ylabel("$y$",rotation = 0, fontsize =10)
plt.legend(loc = "upper left", fontsize=10)

plt.title("Quardratic Predictions Plot")
plt.show()
```
