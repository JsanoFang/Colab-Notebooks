```python
#@title Genshin Pity Calculator Tool -By Jsano
#@markdown ######This Will Assume The Event Banner for Calculations/Pity
#@markdown ---
Number_of_Pity =  8#@param {type:"number"}
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
temp_result=[]
result = []

def calcprob(n, out="", last_digit=1,ones=0):
 
    # if the number becomes nâ€“digit, print it
    if n == 0:
      #if result list is empty or not large enough
      #extend existing table with a longer one
        try:
          temp_result[ones] += 1
        except:
          temp = []
          for x in range (len(temp_result),ones+1,1):
            temp.append(0)
          temp_result.extend(temp)
          temp_result[ones] += 1
        return
 
    # append 1 to the result and recur with one less digit
    calcprob(n - 1, out + '1', 1,ones+1)
 
    # append 0 to the result and recur with one less digit
    # only if the last digit is 1
    if last_digit == 1:
        calcprob(n - 1, out + '0', 0,ones)
 
 





first = 1
second =2
for x in range (1,int(Number_of_Pity),1):
    temp = second; 
    second = first+second;
    first = temp;

#print (second)


calcprob(int(Number_of_Pity))
#print(temp_result)

#transferring temp_result list to a new list while calculating it
for x in range(0,len(temp_result),1):
  result.append((temp_result[x]/second)*100)

#print(result)
#creating figure and adding the subplots
fig = plt.figure()
ax = fig.add_subplot(111)

#setting up the x-axis data, labels , and plotting datas
x = result
plt.title('Number of wanted character Percentage')
plt.xlabel('Number of Event Banner Character got')
plt.ylabel('Percentage (%)')
plt.plot(x, '-o') 
#adding the necessary details needed for the x-axis
plt.xticks(range(0,int(Number_of_Pity)+1)) 
#limiting the graph height to 100%
plt.ylim(0,100);
#adding text for each points on the graph, and limiting the decimal to 2
for i, v in enumerate(result):
    ax.text(i, v+3, "%.2f" %v, ha="center") 


```
